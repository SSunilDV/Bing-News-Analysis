```python
df = spark.read.option("multiline", "true").json("Files/bing-latest-news.json")
# df now is a Spark DataFrame containing JSON data from "Files/bing-latest-technology-news.json".
display(df)
```

```python
df=df.select("value")
```

```python
display(df)
```

```python
from pyspark.sql.functions import explode
df_exploded = df.select(explode(df["value"]).alias('json_object'))
```

```python
display(df_exploded)
```

```python
json_list=df_exploded.toJSON().collect()
```

```python
print(json_list)
```

```python
print(json_list[0])
```

```python
import json

news_json = json.loads(json_list[0])
print(news_json['json_object']['description'])
```

```python
title=[]
description=[]
category=[]
url=[]
image=[]
provider=[]
datePublished=[]

for json_str in json_list:
    try:
        news=json.loads(json_str)
        if news['json_object'].get('category'):
            title.append(news['json_object']['name'])
            description.append(news['json_object']['description'])
            category.append(news['json_object']['category'])
            url.append(news['json_object']['url'])
            #image.append(news['json_object']['image']['thumbnail']['contentUrl'])
            provider.append(news['json_object']['provider'][0]['name'])
            datePublished.append(news['json_object']['datePublished'])
    
    except Exception as e:
        print(f'Error in object {e}' )
```

```python
print(title)
```

```python
from pyspark.sql.types import IntegerType, StringType, StructType, StructField

data = list(zip(title,description,category,url,provider,datePublished))

# Create list of StructField
fields = [StructField("Title", StringType(), True),
StructField("Description", StringType(), True),
StructField("Category", StringType(), True),
StructField("Url", StringType(), True),
StructField("Provider", StringType(), True),
StructField("DatePublished", StringType(), True)
]

# Creating StructType schema
schema = StructType(fields)

# Create DataFrame
df_clean = spark.createDataFrame(data, schema=schema)

# Display DataFrame
df_clean.show()
```

```python
from pyspark.sql.functions import to_date, date_format

#Create a new DataFrame
df_final = df_clean.withColumn("DatePublished", date_format(to_date('DatePublished'),'dd-MM-yyyy'))
display(df_final)
```

```python
from pyspark.sql.utils import AnalysisException

try:
    table='Bing_lake_db.tbl_latest_news'
    df_final.write.format('delta').saveAsTable(table)

except AnalysisException:
    print('Table Already exist')

    df_final.createOrReplaceTempView('vw_df_final')

    spark.sql(f"""  MERGE INTO {table} target_table USING vw_df_final source_view ON source_view.url = target_table.url 
                    WHEN MATCHED AND 
                    source_view.Title <> target_table.Title OR 
                    source_view.Description <> target_table.Description OR
                    source_view.Category <> target_table.Category OR
                    source_view.Url <> target_table.URL OR
                    source_view.Provider <> target_table.Provider OR
                    source_view.DatePublished <> target_table.DatePublished
                    Then UPDATE SET *

                    when not matched then insert *
                    """)

```
